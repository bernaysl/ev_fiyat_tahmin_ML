```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("ev_fiyat_tahmini.csv", sep=";") #dosyadaki veriler noktali virgul ile ayrilmis, bunu belirttik
data.head()
```

```python
data.shape #kac adet verimiz olduguna bakalim
```

```python
x= data["MetreKare"]
y = data["fiyatlar"]
```

```python
plt.scatter(x,y) #dagilim grafigi olusturma

plt.xlabel("Metrekare")
plt.ylabel("fiyat")

plt.show()
```

```python
#regresyon cizgisi olusturacagiz, once modeli kurmamiz gerekiyor
model = LinearRegression() #bu fonksiyon sklearn kutuphesinden geliyor

model.fit(x,y) #x ve y verilerini kullanarak cizgi olustur
```

```python
x = data["MetreKare"].values.reshape((-1,1)) # kac satir oldugunu bilmiyorsak -1 yazariz boylece kendisi hesaplar
x = data["fiyatlar"].values.reshape((-1,1))
```

```python
sns.regplot(x="MetreKare", y="fiyatlar", #cizginin nasil olustugunu gormek istiyorum
            data= data, line_kws={"color": "red"}) # veri setini soyledik ve daha rahat gorbeilmek icin rengi kirmizi yaptik
```

```python
model.predict([[75]]) #koseli parantez icine veri girerek makineden tahmin aliyoruz
```

```python
model.score(x,y) #bu skor 1'e ne kadar yaklasirsa modelimiz o kadar basarilidir.
```
